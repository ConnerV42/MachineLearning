```python
import numpy as np
import matplotlib.pyplot as plt

# We will pick 5 values centered from [0,1] and 5 values centered from [1,0].
# These will be centers of gaussian values of two classes of data.

N = 10
x1, y1 = np.random.multivariate_normal(mean = [0,1], cov = [[0.1, 0], [0, 0.1]], size = N).T
x2, y2 = np.random.multivariate_normal(mean = [1, 0], cov = [[0.1, 0], [0, 0.1]], size = N).T

# gen_data will generate 2 matrices, one for class 0 and another for class 1. It will also return 2 vectors:
# one for each of the labels.s
def gen_data(N):
    x_0 = np.array([np.random.multivariate_normal(mean = [a,b], cov = [[0.2, 0], [0, 0.2]], size = N) for (a,b) in zip(x2, y2)]).reshape(-1, 2)
    x_1 = np.array([np.random.multivariate_normal(mean = [a,b], cov = [[0.2, 0], [0, 0.2]], size = N) for (a,b) in zip(x1, y1)]).reshape(-1, 2)
    return x_0, x_1, [0] * x_0.shape[0], [1] * x_1.shape[0]

N = 100
a0, a1, b0, b1 = gen_data(N)
print(a0.shape)
print(a1.shape)
```

```python
# Plot the data
plt.title("Fake Training Data")
plt.scatter(a0[:,0], a0[:,1], alpha=1, color="b", s=5, label="class 0")
plt.scatter(a1[:,0], a1[:,1], alpha=0.5, color="r", s=5, label="class 1")
plt.legend()

# plot the ideal decision boundary
plt.plot(a0, a0, c="lightblue")
```

```python
# Implement linear classifier (Perceptron Learning Algorithm & Pocket Algorithm)
from matplotlib.colors import ListedColormap

# input vector's xi
x_train = np.concatenate((a0, a1), axis = 0)

# output "answers"
y_train = np.concatenate((b0, b1), axis = 0)

# Convert all 0's in y_train to -1's
y_pla = y_train.copy()
y_pla[y_pla == 0] = -1

def e_in(w, x, y):
    return np.sum(np.abs(np.sign(np.sum(np.multiply(x, w), axis = 1)) - y_pla))

def select_x_y(w, x, y):
    a = np.sign(np.sum(np.multiply(x, w), axis = 1)) - y_pla
    
    # return x and y at the index of the first non-zero value of a
    ind = np.where(a != 0)[0][[0]]
    # print(np.where(a != 0)[0].shape)
    return x[ind], y[ind]
    
def perception_learning_step(w, x, y):
    if(e_in(w, x, y) > 0):
        x_e, y_e = select_x_y(w, x, y)
        w = w + x_e * y_e
    return w
    
def pocket(x, y, N):
    w = np.matrix([1,1])
    w_t = np.random.uniform(0.0, 1.0, 2)
    for t in range(N):
        w_t = perception_learning_step(w_t, x, y)
        if(e_in(w_t, x, y) < e_in(w, x, y)):
            w = w_t
    return w

w = pocket(x_train, y_pla, 1000) #obtain vector W

def gEquation(w, x, y, i):
    result = np.sign(np.dot(w, x[i]))
    y[i] = result
    
g_pla = lambda x, w: np.sign(np.sum(np.multiply(w, x)))    
decisionBound = lambda x_i: -(w[0,0]*x_i)/w[0,1] 

N = 100
a0, a1, b0, b1 = gen_data(N) # Reusing gen_data fn for 

x_test = np.concatenate((a0, a1), axis = 0) # Array of test vectors
y_test = np.empty(2000, dtype=np.int8) # Empty array for results

for i in range(len(x_test)):
    gEquation(w, x_test, y_test, i)

plt.plot(a0, a0, c='red')
plt.scatter(np.squeeze(np.asarray(x_test[:, 0])), np.squeeze(np.asarray(x_test[:, 1])), c=y_test, s=20)
plt.plot(x_test, decisionBound(x_test), c='g')
plt.show()

def mse(w, g, x, y):
    return (1/x.shape[0])*(np.sum((g(x, w) - y)**2))

print("Mean Squared Error: ")
print(mse(w, g_pla, x_train, y_pla))
```

```python
import matplotlib.pyplot as plt
# Implement a KNN classifier
def euclidean(a, b):
    d = a-b
    return np.sqrt(np.sum(d.dot(d)))

def distance_array(pt, x):
    return np.apply_along_axis(lambda z: euclidean(pt, z), 1, x)

# y is assumed to be 0 or 1
def knn(k, pt, x, y):
    idxs = np.argsort(distance_array(pt, x))
    y_ones = sum(y[idxs][:k])
    if y_ones > k - y_ones:
        return 1
    return 0

# predict an array of values
def knn_predict(k, pts, x, y):
    return np.apply_along_axis(lambda p: knn(k, p, x, y), 1, pts)

def decisionBoundary(model, ax):
    x_min, x_max = -0.1, 1.1
    y_min, y_max = -0.1, 1.1
    h = .02 # step size in the mesh
    
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                        np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    
    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    cmap_light = ListedColormap(['#AAAAFF', '#AAFFAA', '#FFAAAA'])
    
    ax.pcolormesh(xx, yy, Z, cmap=cmap_light, alpha=0.5)
    return ax


a0,a1,b0,b1 = gen_data(N)
x_train = np.concatenate((a0, a1), axis=0)
y_train = np.concatenate((b0, b1), axis=0)

k = 3
x_min, x_max = -3, 3
y_min, y_max = -3, 3.5
h = 0.1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

print(xx.shape)
print(yy.shape)
print(np.c_[xx.ravel(), yy.ravel()].shape)

Z = knn_predict(k, np.c_[xx.ravel(), yy.ravel()], x_train, y_train)
Z = Z.reshape(xx.shape)

cmap_light = ListedColormap(['#AAAAFF', '#FFAAAA'])
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
plt.scatter(a0[:,0], a0[:,1], alpha=1, color='b', s=5, label='Class 0')
plt.scatter(a1[:,0], a1[:,1], alpha=1, color='r', s=5, label='Class 1')
plt.legend()
plt.title("K-Nearest Neighbor plot: k = {}".format(k))
plt.show()

g_knn = lambda x, w: np.sign(np.sum(np.multiply(w,x)))
def mse(w, g, x, y):
    return (1/(x.shape[0]))*(np.sum((g(x,w) - y)**2))

print("Mean Squared Error Knn:")
print(mse(w, g_knn, x_train, y_train))


```
